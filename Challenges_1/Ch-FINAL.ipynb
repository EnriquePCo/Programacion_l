Para este challenge final, hare un archivo csv en donde registre diferentes las ventas de diferentes tipos de auto, teniendo el csv listo, lo procesare mediante un programa el cual me muestre las primeras filas, datos estadisticos, tal como las medias, desviacion estandar, minimos y maximos, y por ultimo comparare las ventas de las marcas para despues hacer un barplot, en el cual pueda ver estos resultados graficamente.

```python
import pandas as pd
import matplotlib.pyplot as plt

archivo_csv = 'datos_ventas.csv'
datos_ventas = pd.read_csv(archivo_csv)

print("Primeras filas de los datos:")
print(datos_ventas.head())

resumen_estadistico = datos_ventas.describe()
print(f"\nResumen Estadistico: \n{resumen_estadistico}")

ventas_por_marca = datos_ventas.groupby('Marca')['Ventas'].sum()
print(f"\nVentas por {ventas_por_marca}")

ventas_por_marca.plot(kind='bar', color=['blue','gray','red'], edgecolor='black')
plt.title('Ventas por Marca')
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.show()
```

Conclusion: Como era de esperarse, la marca de auto mas vendida es Nissan, y teniendo esta informacion, se pueden tomar decisiones al respecto.
